```python
#Q5  Apply PCA on X

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Specify the file path where your CSV file is located
file_path = '/Users/sauravkumararya/Downloads/cancerdataset1.csv'

# Load the CSV file into a DataFrame
cancer_df = pd.read_csv(file_path)

# Select columns 1 to 23 as input features (assuming columns are 0-indexed)
input_columns = cancer_df.columns[1:24]  # Columns 1-23 (0-indexed)
X = cancer_df[input_columns]

# Standardize the input features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with desired number of components (e.g., 2 for visualization)
n_components = 2  # Number of principal components
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# Plot the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
plt.bar(range(n_components), explained_variance_ratio, alpha=0.5, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()

# Display the transformed data with reduced dimensions
print(f"Original shape of X: {X.shape}")
print(f"Shape of X after PCA: {X_pca.shape}")
```

```python
#Q7  Apply k-means clustering for this dataset(i.e, X) to form 3 clusters

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Specify the file path where your CSV file is located
file_path = '/Users/sauravkumararya/Downloads/cancerdataset1.csv'

# Load the CSV file into a DataFrame
cancer_df = pd.read_csv(file_path)

# Select columns 1 to 23 as input features (assuming columns are 0-indexed)
input_columns = cancer_df.columns[1:24]  # Columns 1-23 (0-indexed)
X = cancer_df[input_columns]

# Standardize the input features (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensionality (optional for visualization)
n_components = 2  # Number of principal components for visualization
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# Apply k-means clustering with 3 clusters
n_clusters = 3  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Visualize the clusters (using PCA for 2D visualization)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.title(f'k-means Clustering with {n_clusters} Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()

# Assign cluster labels to the original DataFrame
cancer_df['Cluster'] = cluster_labels

# Display the count of samples in each cluster
print(cancer_df['Cluster'].value_counts())
```
