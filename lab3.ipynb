```python
#Q1  Count frequencies of various alphabets (Convert upper case into lower case and input given byuser), 
#     plot the results for this as a bar chart with x-axis being the letter and y-axis as the corresponding frequency.

import matplotlib.pyplot as plt

s=input()
a=s.lower()
data={}
for ch in s:
    data[ch]=a.count(ch)
key=list(data.keys())
value=list(data.values())
plt.bar(key,value,color='purple')
plt.show()
```

```python
#Q4  plot histogram for age of male and female in different plots for the following data of male
#      and female age.
#      male age = [53,51,71,31,33,39,52,27,54,30,64,26,21,54,52,20,59,32]
#     female age = [53,65,68,21,75,46,24,63,61,24,49,41,39,40,25,54,42, 32,48,23,23]


import matplotlib.pyplot as plt
male_age = [53, 51, 71, 31, 33, 39, 52, 27, 54, 30, 64, 26, 21, 54, 52, 20, 59, 32]
female_age = [53, 65, 68, 21, 75, 46, 24, 63, 61, 24, 49, 41, 39, 40, 25, 54, 42, 32, 48, 23, 23]
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(male_age, bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Male Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.hist(female_age, bins=10, color='red', edgecolor='black')
plt.title('Histogram of Female Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()  # Adjust layout to prevent overlappin
plt.show()
```

```python
#Q5 Plot the temperature extremes in certain region of India for each month, starting in January,which are given by (in degrees Celsius).
#   max: 17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18
#   min: -62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58

import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
max= [17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18]
min= [-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58]

plt.plot(months, max, 'r-', label='Max (°C)')
plt.plot(months, min, 'b-', label='Min (°C)')

plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
